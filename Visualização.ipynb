```python
import requests
from bs4 import BeautifulSoup
```

```python
# Aquisitando informações do Site:
resposta = requests.get("https://celke.com.br/artigo/tabela-de-cores-html-nome-hexadecimal-rgb")
resposta
```

```python
beaut_soap = BeautifulSoup(resposta.text,"html.parser");beaut_soap
```

```python
# Extraindo os valores hexadecimais
hexadecimais = []

# Iterando sobre as linhas da tabela
for linha in beaut_soap.find_all('tr')[1:]:  # Pular o cabeçalho
    celulas = linha.find_all('td')
    if len(celulas) > 1:
        valor_hex = celulas[1].get_text(strip=True)  # A segunda célula contém o valor hexadecimal
        hexadecimais.append(valor_hex)
```
